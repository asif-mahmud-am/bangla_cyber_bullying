Remove English language words

```python
import pandas as pd
from nltk.corpus import words
from nltk.tokenize import word_tokenize

def remove_english_words(dataframe, column_name):
    # Download the NLTK words dataset if not already downloaded
    import nltk
    nltk.download('words')

    # Get the set of English words
    english_word_set = set(words.words())

    # Function to check if a word is English
    def is_english(word):
        return word.lower() in english_word_set

    # Function to filter out English words from a text
    def filter_non_english(text):
        tokens = text.split(" ")
        non_english_tokens = [token for token in tokens if not is_english(token)]
        return ' '.join(non_english_tokens)

    # Apply the filter_non_english function to the specified column
    dataframe[column_name] = dataframe[column_name].apply(filter_non_english)

    return dataframe
```

```python
from transformers import BertTokenizer, AutoModelForSequenceClassification , AutoTokenizer
```

Load Tokenizer

```python
tokenizer = AutoTokenizer.from_pretrained("csebuetnlp/banglabert", use_fast=True)
```

Pandas dataframe to dataset for training

```python
from datasets import Dataset
dataset = Dataset.from_pandas(dataset)
```

```python
dataset
```

Tokenization

```python
col_to_delete = ["cleaned"]

def bert_preprocessing_function(examples):
    return tokenizer(examples['cleaned'], truncation=True, max_length=128)

tokenized_datasets = dataset.map(bert_preprocessing_function, batched=True, remove_columns=col_to_delete)
```

```python
tokenized_datasets
```

Train test split

```python
split_dataset = tokenized_datasets.train_test_split(test_size=0.2)
```

```python
train_dataset = split_dataset["train"]
```

```python
test_dataset = split_dataset["test"]
```

Load model

```python
from transformers import AutoModelForSequenceClassification
import torch

model = AutoModelForSequenceClassification.from_pretrained("csebuetnlp/banglabert", num_labels=5, return_dict=True)
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="bert_bangla_bullying",
    learning_rate=1e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=10,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    logging_strategy="epoch",
    # logging_steps=100,
    load_best_model_at_end=True,
)
```

Metrics function

```python
!pip install evaluate
```

```python
import evaluate
import numpy as np
from sklearn import metrics
import torch
import numpy as np

def compute_metrics(eval_pred):


    logits, labels = eval_pred # eval_pred is the tuple of predictions and labels returned by the model
    predictions = np.argmax(logits, axis=-1)

    precision = metrics.precision_score(labels, predictions, average="macro")
    recall = metrics.recall_score(labels, predictions, average="macro")
    f1 = metrics.f1_score(labels, predictions, average="macro")
    accuracy = metrics.accuracy_score(labels, predictions)

    return {"precision": precision, "recall": recall, "f1-score": f1, 'accuracy': accuracy}
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

Train the model

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

trainer.train()
```

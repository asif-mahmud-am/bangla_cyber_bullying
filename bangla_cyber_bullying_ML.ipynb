# **Data Engineering**

```python
df = df[["Description", "Label"]]
```

First, all labels to numeric values

```python
multi_class_label_mapping = {'Political': 0, 'sexual': 1, 'troll': 3, 'Threat': 4, 'Neutral': 5}
binary_class_label_mapping = {'Political': 1, 'sexual': 1, 'troll': 1, 'Threat': 1, 'Neutral': 0}

df['multi_class_label'] = df['Label'].map(multi_class_label_mapping)
df['binary_class_label'] = df['Label'].map(binary_class_label_mapping)
```

```python
df
```

# Let's prepare our data for training [**Remove stopwords, punctuations, apply tf-idf **]

```python
from bnlp import BengaliCorpus as corpus

stopwords = corpus.stopwords
punctuations = corpus.punctuations
```

```python
def preprocess_text(text):
    tokens = nltk.word_tokenize(text)
    tokens = [word.lower() for word in tokens if word not in stopwords and word not in punctuations]
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text
```

```python
df['processed_text'] = df['Description'].apply(preprocess_text)
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(lowercase=False, ngram_range=(2,2))

tfidf_matrix = tfidf_vectorizer.fit(df['processed_text'].to_list())
```

Taking the input and output features

```python
X = tfidf_matrix.transform(df["processed_text"].to_list()).toarray()
y_bin = df["binary_class_label"]
y_mul = df["multi_class_label"]
```

Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train_bin, y_test_bin = train_test_split(X, y_bin, test_size = 0.2, random_state = 42)  #binary classification
```

```python
X_train, X_test, y_train_mul, y_test_mul = train_test_split(X, y_mul, test_size = 0.2, random_state = 42) #multiclass classificcation
```

# Model Training

Binary Classification with Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
dt = DecisionTreeClassifier(max_leaf_nodes=8)
```

```python
dt.fit(X_train, y_train_bin)
```

```python
y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_pred, y_test_bin))
```

We can see that for minority class, the performance is not good, because the dataset is imbalanced. So we will apply SMOTE technique to balance the dataset and train model again.

```python
y_bin.value_counts()
```

```python
from imblearn.under_sampling import ClusterCentroids
cc = ClusterCentroids(random_state=0)
X_under, y_under = cc.fit_resample(X, y_bin)
```

```python
X_train_under, X_test_under, y_train_under, y_test_under = train_test_split(X_under, y_under, test_size = 0.2, random_state = 0)
X_train_under.shape, y_train_under.shape, X_test_under.shape, y_test_under.shape
```

```python
dt.fit(X_train_under, y_train_under)
```

```python
y_pred = dt.predict(X_test_under)
print(classification_report(y_pred, y_test_under))
```
